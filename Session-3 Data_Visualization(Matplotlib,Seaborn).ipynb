## Bivariate Data

### a.Line Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7]
y = [i**2 for i in x]
```

```python
y
```

```python
# Draw Line Chart
plt.plot(x,y, marker='H',markersize=8,
         markerfacecolor='orange',
        markeredgecolor='midnightblue',
        linestyle='dotted')
# Define label
plt.title('Line Chart',fontsize=20)
plt.xlabel('X - Values')
plt.ylabel('Y - Values')
plt.show()
```

## Categorical Vs Numerical Data

```python
courses = ['Data Science','Data Analysis','Mysql','Tableau',"Powerbi"]
students = [120,178,65,100,132]
```

```python
# Set the Figure Size
plt.figure(figsize=(8,4))

# Plot the barchart
plt.bar(courses,students,color=['r','b','g','y','c'],
       alpha=0.5,
       edgecolor='black')

plt.xticks(fontsize=10,rotation=45)

# Define Label
plt.title('Student Count',fontsize=18)
plt.xlabel('Courses',color='blue')
plt.ylabel('No.of Students',color='blue')
plt.show()
```

### Horizontal Barchart

```python
# Set the Figure Size
plt.figure(figsize=(8,4))

# Plot the barchart
plt.barh(courses,students,
         color=['r','b','g','y','c'],
       alpha=0.5,
       edgecolor='black',
        linewidth=1)

plt.xticks(fontsize=10,rotation=45)

# Define Label
plt.title('Student Count',fontsize=18,color='midnightblue')
plt.xlabel('Courses',color='blue')
plt.ylabel('No.of Students',color='blue')
plt.show()
```

### c. Pie Chart

```python
plt.pie(students,labels=courses,
       autopct='%0.1f%%',
       textprops={'fontsize':12,'color':'black'},
       wedgeprops={'edgecolor':'white','linewidth':3},
       explode=[0,0,0.3,0,0],
       colors = ['r','b','g','y','c'])
plt.show()
```
